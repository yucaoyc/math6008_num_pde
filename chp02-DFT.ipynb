```python
from scipy.fft import fft, ifft
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 80

# rescale integer n by N
n = np.linspace(0, 1, N+1)[0:N]
a = 8 # frequency

# mixed frequencies
f1 = np.cos(2*np.pi*a*n) + complex(0,1)*np.sin(2*np.pi*a*n)
f2 = np.cos(2*np.pi*a/2*n) + complex(0,1)*np.sin(2*np.pi*a/2*n)
f = f1

plt.figure(figsize=(6,2.5))
plt.subplot(1,2,1)
plt.plot(n, f.real, ls='-', label="real")
plt.plot(n, f.imag, ls='--', label="imag")
plt.xlabel(r"$n/N$", fontsize=14);
plt.title(r'$f_n$', fontsize=14)
plt.legend()

plt.subplot(1,2,2)
plt.plot(np.array(range(N))/N, fft(f).real, 'o', label="real")
plt.plot(np.array(range(N))/N, fft(f).imag, label="imag")
plt.title(r"$\hat{f}_k$", fontsize=14)
plt.xlabel(r"$k/N$", fontsize=14);
plt.legend()

plt.tight_layout()
```

```python
# with mixed frequencies
f = 2*f1 + f2

plt.figure(figsize=(6,2.5))
plt.subplot(1,2,1)
plt.plot(n, f.real, ls='-', label="real")
plt.plot(n, f.imag, ls='--', label="imag")
plt.xlabel(r"$n/N$", fontsize=14);
plt.title(r'$f_n$', fontsize=14)
plt.legend()

plt.subplot(1,2,2)
plt.plot(np.array(range(N))/N, fft(f).real, 'o', label="real")
plt.plot(np.array(range(N))/N, fft(f).imag, label="imag")
plt.title(r"$\hat{f}_k$", fontsize=14)
plt.xlabel(r"$k/N$", fontsize=14);
plt.legend()

plt.tight_layout()
```
